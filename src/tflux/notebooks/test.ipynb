```python
import matplotlib.pyplot as plt
from codetiming import Timer
from tflux.utils.logging import get_logger
from pathlib import Path
from tflux.pipeline import config, run
from tflux.io import paths
```

```python
logger = get_logger(__name__)
```

```python
# Default output path: tflux/outputs/<date_and_index>/
root = paths.find_root()
data_dir_path, output_dir_path = paths.prepare_io(set_data_dir_path=config.DATA_DIR_PATH, set_output_dir_path=None, include_root=True)

# Save config to .txt
if config.save_config:
    with open(Path(root / "src/tflux/pipeline/config.py"), "r") as src, open(output_dir_path / "config.txt", "w") as out:
        out.writelines(src.readlines())
```

```python
if config.load_pkl:
    with open(config.pkl_path, "rb") as f:
        logger.info(f"Loading sample from pickle file: {config.pkl_path}")
        sample = pickle.load(f)
```

```python
from tflux.io.png_to_pdf import pngs_to_pdf
from tflux.plotting.points import plot_cell_3d_with_norms

cell_dir = output_dir_path / "cells2"
for cell in sample.cells:
    # fig = plot_cell_3d(cell=cell, title=cell)
    fig = plot_cell_3d_with_norms(cell=cell, title=cell)
    png_name = f'C{cell.cell_index}.png'
    fig.savefig(cell_dir / png_name)
    plt.close(fig)

pngs_to_pdf(input_dir=cell_dir, output_path=Path(cell_dir / "cells.pdf"))
```
